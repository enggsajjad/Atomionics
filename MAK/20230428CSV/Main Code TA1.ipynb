```python
# list to store file lines
lines = []
# read file
with open(r"25.04.2023, 08.18,  958.77349 nm.lta", 'r') as fp:
    # read an store all lines into list
    lines = fp.readlines()

# Write file
with open(r"25.04.2023, 08.18,  958.77349 nm_modified.lta", 'w') as fp:
    # iterate each line
    for number, line in enumerate(lines):
        # delete line 5 and 8. or pass any Nth line you want to remove
        # note list index starts from 0
        if number not in range(56):
            fp.write(line)
```

```python
dataset = pd.read_csv('25.04.2023, 08.18,  958.77349 nm_modified.lta', sep='\s+',header = None)
accData=dataset.values.tolist()
#accData=np.asarray([np.asarray(elem) for elem in accData])
dataset['A'] = dataset[0] 
dataset['B'] = dataset[1] 

C1 = np.array(dataset['A'])
C2 = np.array(dataset['B'])
C1 = C1.astype(float)
C2 = C2.astype(float)

#select first 100 values
C2_100 = C2[0:100]
C1_100 = C1[0:100]
```
