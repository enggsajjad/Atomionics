The Transition probability is given by $P=A(1-C\cos\Phi)$. <br>
Here $\Delta P=AC\sin\Phi\Delta\Phi$ and $\Phi=k_{eff}gT^{2}$<br>
$\sigma_{\Phi}=\Delta P/AC$<br>
$\sigma_{\Phi}=k_{eff}g\sin\Phi\Delta\cos\theta T^{2}$, here $\theta$ is the tilt angle.<br>
$\sigma_{\Phi}=k_{eff}g\sin\Phi(1-\cos\theta)T^{2}$<br>
$\sigma_{g}=\sigma_{\Phi}/(k_{eff}T^2)$

```python
b=1.03
tiltAngle=b**np.linspace(1,200,200)*5*10**(-3)*np.pi/180
Sigma_g=-g0*np.sin(keff*g0*T**2)*(1-np.cos(tiltAngle))
#print(tiltAngle)
```

```python
import matplotlib.pyplot as plt
figure,ax=plt.subplots(1)
#ax.set_xscale('log')
#ax.set_yscale('log')
#plt.yscale('symlog')
ax.plot(tiltAngle*10**3,Sigma_g*10**8)
ax.set_xlabel('tilt angle in mrads',fontsize=15)
ax.set_ylabel(r'$\delta g$ in $\mu$ gals',fontsize=15)
ax.tick_params(labelsize=15)
```


If the setup is tilted at angle angle theta with respect to the gravity field vertical direction the effective $g_{Eff}=g(1-\cos\theta)$ is experienced by the setup. For the tilt value $\sigma_{\theta}$ the change in $g$ that is $\delta g$ is obtained above x axis in m rads and y axis is in micro gals.

```python
#sigma_g=-g0*np.sin(keff*g0*T*T)*(np.pi/180)*stdDevTheta*stdDevTheta

sigma_g=(keff*g0*T*T)*(np.pi/180)*stdDevTheta*stdDevTheta

figures,ax=plt.subplots(1)
#ax.set_xscale('log')
#ax.set_yscale('log')
#plt.yscale('symlog')
ax.plot(stdDevTheta*10**6,sigma_g*10**8)
ax.set_xlabel('tilt angle in microrads',fontsize=15)
ax.set_ylabel(r'$\delta g$ in $\mu$ gals',fontsize=15)
ax.tick_params(labelsize=15)
```
