## Point Mass<br>
we derived the vertical component of the gravitational acceleration due to a point mass with mass m as:

$\Delta_g = \frac {Gmz}{(x^2+z^2)^{3/2}}$


```python
#Assuming a ponit mass 
N = 1
start = 1000
G = 6.76e-11
x = np.linspace(-(start),(start),(start*2+1))
V = 10 * 4 *1000
Rho = 1550
M = V * Rho
print ("Point Mass Volume: ",V)
print ("Point Mass Mass: ",M)
m = M
Zi = 175             
di = 0            

delta_g=0.0*x
for tau in x:
    delta_g[x==tau] = (G*m*Zi / (((tau-di)**2+Zi**2)**(3/2)))

fig, ax = plt.subplots()
ax.plot(x,delta_g*(10**8))
ax.set_xlabel('distance (x)',fontsize=15)
ax.set_ylabel('Gravitational Accel.$\Delta_g$',fontsize=15)
ax.set_title('Ponit Mass Gravitational Acceleration')
plt.text( 1.0, 1.0, 'N=%s\n$\Delta_g$=%s'%(N,round(max(delta_g*(10**8)),2)), ha='right', va='top',  transform=ax.transAxes)
```

## Complex Body with Multiple Point Masses<br>
We can approximate the body with complex shape as a distribution of point masses. 
The gravitational attraction of the body is then nothing more than the sum of the gravitational attractions of all of 
the individual point masses as illustrated below.
If there are n point masses, this equation can be written more compactly in mathematical notation, this sum can be written as:

$\Delta_g = \sum_{i=0}^n \frac {Gmz_i}{((x-d_i)^2+z_i^2)^{3/2}}$

where z represents the depth of burial of each point mass, d represents the horizontal position of each point mass, 
and x represents the horizontal position of the observation point. Only the first three terms have been written in 
this equation. There is, in actuality, one term in this expression for each point mass.

```python
import numpy as np
import matplotlib.pyplot as plt

dRock = 175
wRock = 4
lRock = 1000
diffDepth = 185-175

start = 5000

N = 64
n = (int)((N)**(0.5)) # this will be the square root of the n

Rho = 1550
V = 4 * 10 * 1000
print ("Whole Mass Volume: ",V)
M = V * Rho
print ("Whole Mass Mass: ",M)
G = 6.76e-11
x = np.linspace(-(start),(start),(start*2+1))
v = V/(n**2)
print ("Point Mass Volume: ",v)
m = 1#v *Rho
#m = M/(n**2) # also
print ("Point Mass Mass: ",m)

Zi = np.random.rand(n,1)*diffDepth+dRock
#Zi = np.linspace(0,1,n)*diffDepth+dRock
di = np.random.rand(n,1)*lRock-((int)(lRock/2))
#di = np.linspace(0,1,n)*lRock-((int)(lRock/2))

delta_g=0.0*x
delta_g1=0.0*x
delta_g2=0.0*x
delta_g3=0.0*x
delta_g4=0.0*x
delta_g5=0.0*x
delta_g6=0.0*x
delta_g7=0.0*x
delta_g8=0.0*x
delta_g0=0.0*x
#Calculate delta_g for all mass points
s1 = np.random.rand(1,1)*(np.size(Zi))
s1 = (int)(np.floor(s1))
s2 = np.random.rand(1,1)*(np.size(di))
s2 = (int)(np.floor(s2))

for tau in x:
    sum = 0;
    #for td in di:
    for td in di[s1::8]:
        #for tz in Zi:
        for tz in Zi[s2::8]:
            sum  = sum + (G*m*tz / ((((tau-td)**2)+(tz**2))**(3/2)))
    delta_g[x==tau] = sum
    #delta_g[x==tau] = np.sum(G*m*Zi / ((((tau-di)**2)+(Zi**2))**(3/2)))

#Calculate delta_g for some sample mass points
for tau in x:
    delta_g1[x==tau] = (G*m*Zi[1] / (((tau-di[1])**2+Zi[1]**2)**(3/2)))
    delta_g2[x==tau] = (G*m*Zi[int(n/8)] / (((tau-di[int(n/8)])**2+Zi[int(n/8)]**2)**(3/2)))
    delta_g3[x==tau] = (G*m*Zi[int(n/4)] / (((tau-di[int(n/4)])**2+Zi[int(n/4)]**2)**(3/2)))
    delta_g4[x==tau] = (G*m*Zi[3*(int(n/8))] / (((tau-di[3*(int(n/8))])**2+Zi[3*(int(n/8))]**2)**(3/2)))
    delta_g5[x==tau] = (G*m*Zi[int(n/2)] / (((tau-di[int(n/2)])**2+Zi[int(n/2)]**2)**(3/2)))
    delta_g6[x==tau] = (G*m*Zi[int(5*n/8)] / (((tau-di[int(5*n/8)])**2+Zi[int(5*n/8)]**2)**(3/2)))
    delta_g7[x==tau] = (G*m*Zi[int(3*n/4)] / (((tau-di[int(3*n/4)])**2+Zi[int(3*n/4)]**2)**(3/2)))
    delta_g8[x==tau] = (G*m*Zi[int(7*n/8)] / (((tau-di[int(7*n/8)])**2+Zi[int(7*n/8)]**2)**(3/2)))
    delta_g0[x==tau] = (G*m*Zi[int(n-1)] / (((tau-di[int(n-1)])**2+Zi[int(n-1)]**2)**(3/2)))

plt.plot(x,delta_g0*(10**8))
plt.plot(x,delta_g1*(10**8))
plt.plot(x,delta_g2*(10**8))
plt.plot(x,delta_g3*(10**8))
plt.plot(x,delta_g4*(10**8))
plt.plot(x,delta_g5*(10**8))
plt.plot(x,delta_g6*(10**8))
plt.plot(x,delta_g7*(10**8))
plt.plot(x,delta_g8*(10**8))

plt.xlabel('distance (x)',fontsize=15)
plt.ylabel('Gravitational Accel.$\Delta_g$',fontsize=15)
plt.title('Multiple Ponit Mass Gravitational Acceleration')
plt.text( 1.0, 1.0, 'N=%s\n$\Delta_{g0}$=%s\n$\Delta_{g1}$=%s\n$\Delta_{g2}$=%s\n$\Delta_{g3}$=%s\n$\Delta_{g4}$=%s\n$\Delta_{g5}$=%s\n$\Delta_{g6}$=%s\n$\Delta_{g7}$=%s\n$\Delta_{g8}$=%s'%(N,round(max(delta_g0*(10**8)),2),round(max(delta_g1*(10**8)),2),round(max(delta_g2*(10**8)),2),round(max(delta_g3*(10**8)),2),round(max(delta_g4*(10**8)),2),round(max(delta_g5*(10**8)),2),round(max(delta_g6*(10**8)),2),round(max(delta_g7*(10**8)),2),round(max(delta_g8*(10**8)),2)), ha='right', va='top',  transform=ax.transAxes)
```
