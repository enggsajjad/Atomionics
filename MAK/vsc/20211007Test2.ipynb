```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import csv
```

```python
snr = pd.read_csv('Phi_vib.csv')
accData=snr.values.tolist()
accData=np.asarray([np.asarray(elem) for elem in accData])
snr['vib'] = snr['xdft']
PhiVib = np.array(snr['vib'])
m = np.linspace(0,np.size(PhiVib),np.size(PhiVib))

A = 1
C = 1
#Keff = 1
Keff = 0
g = 9.8
#PhiVib=1
#Phi=1
Phi=np.pi/2
T=1
alpha1=0#np.linspace(0,360,10000)
#phf=57*10**(-3)  #Phase fluctuation
phf=np.linspace(np.pi/2, 3*np.pi/2, np.size(m))
#phf=np.linspace(np.pi/2, 3*np.pi/2, 1000)

P1 = A- C/2 * np.cos(( Keff*g - alpha1 ) * T**2 + PhiVib + Phi+phf)#*(np.pi/180)

figure,ax=plt.subplots(1)
ax.plot(m, P1,'r')
ax.set_xlabel(r'measurements',fontsize=25)
ax.set_ylabel(r'Atomic Phase shift(rad)',fontsize=25)

plt.xlim(0, 300)
plt.ylim(1, 2)
#print(PhiVib)
```
